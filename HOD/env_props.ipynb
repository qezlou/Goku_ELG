```python
import h5py
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 1, 1000, endpoint=False)
X, Y = np.meshgrid(x, x, indexing='ij')
with h5py.File(f'/home/qezlou/HD2/HETDEX/data/potential/q_maps.hdf5', 'r') as f:
    for sigma in ['1.0','1.5','2.0','2.5']:
        q = f[f'Shear-s{sigma}']
        print(q.shape)
        plt.contourf(X[:, :], Y[:, :], q[:, :, 10], 20, cmap='viridis', vmin=0, vmax=5)
        plt.colorbar()
        plt.title(f'Q, sigma={sigma}')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
```

```python
from scipy.ndimage import gaussian_filter as gf
with h5py.File('/home/qezlou/HD2/HETDEX/data/density/full_density.hdf5','r') as f:
    dens = f['DM/dens'][:]
    dens = gf(dens, sigma=4, mode='wrap')

    plt.imshow(dens[500,:,:], cmap='hot', vmin=0, vmax=5)
    plt.colorbar()
    plt.title('den[500,:,:]')
    plt.show()
```
