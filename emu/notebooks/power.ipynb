## Check the power:

```python
import h5py
from os import path as op
import numpy as np
from matplotlib import pyplot as plt

data_dir = '/home/qezlou/HD2/HETDEX/cosmo/data/power/'

def plot_power(data_file):
    with h5py.File(data_file, 'r') as f:
        k = f['k'][:]
        #mean = np.mean(f['pk'][:], axis=0)
        #std = np.std(f['pk'][:], axis=0)
        pk = f['pk'][:]

    fig, ax = plt.subplots(figsize=(12, 6))
    #_ = ax.errorbar(k, mean, std)
    ax.plot(k, pk.T, alpha=0.4)

    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(0.2, 1e0)
    ax.set_ylim(3e5, 3e6)


plot_power(op.join(data_dir, 'Zheng07_seeds_cosmo_10p_Box1000_Part750_0040.hdf5'))
plot_power(op.join(data_dir, 'Zheng07_seeds_compressed_10p_Box250_Part750_0316_mesh_res_0.25.hdf5'))
plot_power(op.join(data_dir, 'Zheng07_seeds_compressed_10p_Box250_Part750_0316_mesh_res_1.0.hdf5'))
```
