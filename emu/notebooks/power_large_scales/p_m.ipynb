```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
from scipy.interpolate import interp1d
```

```python
import importlib
from gal_goku import summary_stats
importlib.reload(summary_stats)

def plot_phh_mm(fid='HF', sample=21):

    data_dir='/home/qezlou/HD2/HETDEX/cosmo/data'
    halo_pow = summary_stats.HaloHaloPower(data_dir, fid=fid, z=2.5)
    mat_pow = summary_stats.MatterPower(data_dir, fid=fid, log_k_interp=np.log10(halo_pow.k), z=2.5)

    fig, ax = plt.subplots(1,3, figsize=(15, 5))
    mpairs = np.array([11.5, 11.5])
    sim_ind = np.random.choice(range(halo_pow.pk.shape[0]), sample, replace=False)
    for sim in sim_ind:
        ind_mbin = np.where(np.isclose(halo_pow.mpairs[:,0], mpairs[0], atol=1e-3) & np.isclose(halo_pow.mpairs[:,1], mpairs[1], atol=1e-3))[0]
        ax[0].plot(halo_pow.k, halo_pow.pk[sim,ind_mbin,:].squeeze(), label='Matter Power Spectrum', color='blue')
        ax[0].plot(mat_pow.k, mat_pow.pk[sim,:].squeeze(), label='Halo-Halo Power Spectrum', color='orange')
        ax[1].plot(mat_pow.k, halo_pow.pk[sim,ind_mbin,:].squeeze() / mat_pow.pk[sim,:].squeeze(), label='P_hh / P_m', color='green')
        ax[2].plot(mat_pow.k, np.log( halo_pow.pk[sim,ind_mbin,:].squeeze() / mat_pow.pk[sim, :].squeeze()), color='red')


    ax[0].set_xscale('log')
    ax[0].set_yscale('log')
    ax[1].set_ylim((0, 200))
    ax[1].set_ylabel('P_hh / P_m')
    ax[1].set_xlabel('k')

    ax[2].set_ylabel(r'$log(P_{hh} / P_{mm})$')
    ax[2].set_xlabel('k')
    ax[2].set_ylim(0,7)

```

## Compare fidelites:

```python
import importlib
from gal_goku import summary_stats
importlib.reload(summary_stats)
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def find_common_tags(tag1, tag2):
    """
    Find common tags between two lists of tags.
    """
    sim_ids = [int(a[-4::]) for a in tag1]
    sim_ids2 = [int(a[-4::]) for a in tag2]
    ind1 = np.where(np.isin(sim_ids, sim_ids2))[0]
    ind2 = np.where(np.isin(sim_ids2, sim_ids))[0]
    return ind1, ind2

def plot_common_phh_mm(sim=20, mpairs=np.array([11.5, 11.5])):

    data_dir='/home/qezlou/HD2/HETDEX/cosmo/data'
    halo_pow_hf = summary_stats.HaloHaloPower(data_dir, fid='HF', z=2.5)
    mat_pow_hf = summary_stats.MatterPower(data_dir, fid='HF', log_k_interp=np.log10(halo_pow_hf.k), z=2.5)

    halo_pow_l2 = summary_stats.HaloHaloPower(data_dir, fid='L2', z=2.5)
    mat_pow_l2 = summary_stats.MatterPower(data_dir, fid='L2', log_k_interp=np.log10(halo_pow_l2.k), z=2.5)

    ind_hf, ind_l2 = find_common_tags(halo_pow_hf.sim_tags, halo_pow_l2.sim_tags)
    print(f'Found {len(ind_hf)} common simulations between HF and L2.')

    
    ind_mbin = np.where(np.isclose(halo_pow_hf.mpairs[:,0], mpairs[0], atol=1e-3) & np.isclose(halo_pow_hf.mpairs[:,1], mpairs[1], atol=1e-3))[0]
    print(ind_mbin)

    fig, ax = plt.subplots(len(ind_hf),3, figsize=(14, 4*len(ind_hf)))
    for i, sim in enumerate(ind_hf):
        print(halo_pow_hf.pk[sim,ind_mbin,:].shape, sim ,ind_mbin, ind_hf)
        ax[i, 0].plot(halo_pow_hf.k, halo_pow_hf.pk[sim,ind_mbin,:].squeeze(), label='Matter Power Spectrum', color='C0')
        ax[i, 0].plot(halo_pow_l2.k, halo_pow_l2.pk[ind_l2[i],ind_mbin,:].squeeze(), label='Halo-Halo Power Spectrum (L2)', color='C0', linestyle='--')
        ax[i, 0].plot(mat_pow_hf.k, mat_pow_hf.pk[sim,:].squeeze(), label='Halo-Halo Power Spectrum', color='C2')
        ax[i, 0].plot(mat_pow_l2.k, mat_pow_l2.pk[ind_l2[i],:].squeeze(), label='Matter Power Spectrum (L2)', color='C2', linestyle='--')

        log_ratio_hf = np.log( halo_pow_hf.pk[sim,ind_mbin,:].squeeze() / mat_pow_hf.pk[sim, :].squeeze())
        ax[i,1].plot(mat_pow_hf.k, log_ratio_hf, color='C0')
        log_ratio_l2 = np.log( halo_pow_l2.pk[ind_l2[i],ind_mbin,:].squeeze() / mat_pow_l2.pk[ind_l2[i], :].squeeze())
        ax[i,1].plot(mat_pow_l2.k, log_ratio_l2, color='C0', linestyle='--')
        
        # plot the  the difference between the two log ratios
        ratio_hf_interp = interp1d(np.log10(mat_pow_hf.k), log_ratio_hf, bounds_error=False, fill_value='extrapolate')(np.log10(mat_pow_l2.k))
        ax[i,2].plot(mat_pow_l2.k, ratio_hf_interp - log_ratio_l2, color='C0')
        

        ax[i, 0].set_xscale('log')
        ax[i, 0].set_yscale('log')
        ax[i,0].set_yscale('log') 

        ax[i, 1].set_ylabel(r'$log(P_{hh} / P_{mm})$')
        ax[i, 1].set_xlabel('k')
        ax[i, 1].set_ylim(1, 6)

        ax[i,2].set_xlim(0,0.5)
        ax[i,2].set_ylim(-0.2, 0.2)
        ax[i,2].grid()

plot_common_phh_mm()
```

## BAO in the original HF sims:

```python
import importlib
from gal_goku import summary_stats
importlib.reload(summary_stats)
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
import mcfit

def find_common_tags(tag1, tag2):
    """
    Find common tags between two lists of tags.
    """
    sim_ids = [int(a[-4::]) for a in tag1]
    sim_ids2 = [int(a[-4::]) for a in tag2]
    ind1 = np.where(np.isin(sim_ids, sim_ids2))[0]
    ind2 = np.where(np.isin(sim_ids2, sim_ids))[0]
    return ind1, ind2

def plot_common_phh_mm(sim=20, mpairs=np.array([11.0, 11.0])):

    log_k_interp = np.log10(np.logspace(np.log10(7e-3), np.log10(0.5), num=300))
    data_dir='/home/qezlou/HD2/HETDEX/cosmo/data'
    halo_pow_hf = summary_stats.HaloHaloPower(data_dir, fid='HF', z=2.5, log_k_interp=log_k_interp)
    mat_pow_hf = summary_stats.MatterPower(data_dir, fid='HF', log_k_interp=log_k_interp, z=2.5)
    halo_pow_l2 = summary_stats.HaloHaloPower(data_dir, fid='L2', z=2.5, log_k_interp=log_k_interp)
    mat_pow_l2 = summary_stats.MatterPower(data_dir, fid='L2', log_k_interp=log_k_interp, z=2.5)

    ind_hf, ind_l2 = find_common_tags(halo_pow_hf.sim_tags, halo_pow_l2.sim_tags)
    print(f'Found {len(ind_hf)} common simulations between HF and L2.')

    
    ind_mbin = np.where(np.isclose(halo_pow_hf.mpairs[:,0], mpairs[0], atol=1e-3) & np.isclose(halo_pow_hf.mpairs[:,1], mpairs[1], atol=1e-3))[0]
    print(ind_mbin)

    fig, ax = plt.subplots(len(ind_hf),3, figsize=(14, 4*len(ind_hf)))
    for i, sim in enumerate(ind_hf):
        print(halo_pow_hf.pk[sim,ind_mbin,:].shape, sim ,ind_mbin, ind_hf)
        ax[i, 0].plot(halo_pow_hf.k, halo_pow_hf.pk[sim,ind_mbin,:].squeeze(), label='Matter Power Spectrum', color='C0')
        ax[i, 0].plot(halo_pow_l2.k, halo_pow_l2.pk[ind_l2[i],ind_mbin,:].squeeze(), label='Halo-Halo Power Spectrum (L2)', color='C0', linestyle='--')
        ax[i, 0].plot(mat_pow_hf.k, mat_pow_hf.pk[sim,:].squeeze(), label='Halo-Halo Power Spectrum', color='C2')
        ax[i, 0].plot(mat_pow_l2.k, mat_pow_l2.pk[ind_l2[i],:].squeeze(), label='Matter Power Spectrum (L2)', color='C2', linestyle='--')

        rhh_hf, xi_hh_hf = mcfit.P2xi(halo_pow_hf.k, l=0, lowring=True)(np.copy(halo_pow_hf.pk[sim,ind_mbin,:]).squeeze(), extrap=True)
        rhh_l2, xi_hh_l2 = mcfit.P2xi(halo_pow_l2.k, l=0, lowring=True)(np.copy(halo_pow_l2.pk[ind_l2[i],ind_mbin,:]).squeeze(), extrap=True)
        ax[i,1].plot(rhh_hf, rhh_hf * xi_hh_hf, label='Halo-Halo Correlation Function (HF)', color='C0')
        ax[i,1].plot(rhh_l2, rhh_l2 * xi_hh_l2, label='Halo-Halo Correlation Function (L2)', color='C0', linestyle='--')

        rhh_hf, xi_hh_hf = mcfit.P2xi(mat_pow_hf.k, l=0, lowring=True)(np.copy(mat_pow_hf.pk[sim,:]).squeeze(), extrap=True)
        rhh_l2, xi_hh_l2 = mcfit.P2xi(mat_pow_l2.k, l=0, lowring=True)(np.copy(mat_pow_l2.pk[ind_l2[i],:]).squeeze(), extrap=True)
        ax[i,2].plot(rhh_hf, rhh_hf * xi_hh_hf, label='Matter-Halo Correlation Function (HF)', color='C2')
        ax[i,2].plot(rhh_l2, rhh_l2 * xi_hh_l2, label='Matter-Halo Correlation Function (L2)', color='C2', linestyle='--')


        ax[i, 0].set_xscale('log')
        ax[i, 0].set_yscale('log')
        ax[i,0].set_yscale('log') 

        ax[i, 1].set_xlabel('k')
        ax[i, 1].set_xlim((10, 200))
        ax[i,1].set_ylim(0,1)
        ax[i,2].set_xlim(10, 200)
        ax[i,2].set_ylim(0, 0.05)


plot_common_phh_mm()
```
